```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset (replace 'student_dataset.csv' with your actual file name)
df = pd.read_csv('/content/student_data_250.csv')

# 1. Histogram Analysis - GPA Distribution
plt.figure(figsize=(8, 5))
sns.histplot(df['GPA'], bins=10, kde=True, color='blue')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.title('Distribution of GPA')
plt.show()

# 2. Scatterplot of GPA vs. Attendance Rate
plt.figure(figsize=(8, 5))
sns.regplot(x=df['Attendance Rate (%)'], y=df['GPA'], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.xlabel('Attendance Rate (%)')
plt.ylabel('GPA')
plt.title('Scatterplot of GPA vs. Attendance Rate')
plt.show()

# 3. Bar Chart - Number of Students per Major
plt.figure(figsize=(10, 5))
sns.countplot(y=df['Major'], order=df['Major'].value_counts().index, palette='viridis')
plt.xlabel('Number of Students')
plt.ylabel('Major')
plt.title('Number of Students per Major')
plt.show()

# 4. Grouped Bar Chart - Extracurricular Activities Across Graduation Years
plt.figure(figsize=(10, 5))
sns.countplot(x=df['Graduation Year'], hue=df['Extracurricular Activities'], palette='coolwarm')
plt.xlabel('Graduation Year')
plt.ylabel('Number of Students')
plt.title('Extracurricular Activities Participation by Graduation Year')
plt.legend(title='Extracurricular Activities')
plt.show()

# Answering the Analysis Questions
print("Most common GPA range:", df['GPA'].mode()[0])
print("Major with the highest student enrollment:", df['Major'].value_counts().idxmax())
print("Correlation between GPA and Attendance Rate:", df[['GPA', 'Attendance Rate (%)']].corr().iloc[0, 1])

# Checking if students with extracurricular activities have higher attendance rates
avg_attendance = df.groupby('Extracurricular Activities')['Attendance Rate (%)'].mean()
print("Average Attendance Rate by Extracurricular Activities:\n", avg_attendance)
```
